```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

!wget "https://t4.ftcdn.net/jpg/00/02/65/21/360_F_2652182_kzshWTrp8SC1KwkMbLnt0NCVtD1XUf.jpg"

image = cv2.imread('360_F_2652182_kzshWTrp8SC1KwkMbLnt0NCVtD1XUf.jpg')  # ใส่ path ของภาพที่ต้องการ
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # เปลี่ยนจาก BGR (ที่ cv2 ใช้) เป็น RGB

plt.imshow(image)

plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
import cv2

!wget "https://www.publicdomainpictures.net/pictures/240000/nahled/checkerboard-with-ball-shape.jpg"

image = cv2.imread('checkerboard-with-ball-shape.jpg',cv2.IMREAD_GRAYSCALE)
image.shape

image

```

```python
# แปลงเป็น black and white ด้วย threshold
_, binary_image = cv2.threshold(image, 127, 1, cv2.THRESH_BINARY)

# บีบอัดภาพให้เป็น 1 bit ต่อ pixel
h, w = binary_image.shape
# ตรวจสอบว่าขนาดของภาพหารด้วย 8 ลงตัวหรือไม่
if w % 8 != 0:
    # ถ้าไม่ลงตัวให้เติม padding (ค่าของ padding เป็น 0)
    padding = 8 - (w % 8)
    binary_image = np.pad(binary_image, ((0, 0), (0, padding)), 'constant', constant_values=0)

# สร้างอาร์เรย์สำหรับเก็บค่าที่บีบอัดแล้ว
compressed_image = np.packbits(binary_image, axis=1)

compressed_image.shape

compressed_image
```

```python
h = 615  # ความสูงของภาพ (จำนวน pixel)
w = 615  # ความกว้างของภาพ (จำนวน pixel)

# คำนวณขนาดของภาพหลังจาก padding (ถ้ามี padding ตอนบีบอัด)
if w % 8 != 0:
    padded_w = w + (8 - w % 8)  # ขนาดของภาพหลัง padding
else:
    padded_w = w

# แปลงข้อมูลที่บีบอัดกลับเป็น array ของบิต
binary_image = np.unpackbits(compressed_image).reshape(h, padded_w)

# ตัดส่วน padding ออกถ้ามี (ถ้าภาพต้นฉบับมีการเติม padding)
binary_image = binary_image[:, :w]  # ตัดเฉพาะส่วนของภาพจริงออกมา

# แปลงจาก 0 และ 1 กลับไปเป็นภาพขาวดำ (0 และ 255)
reconstructed_image = binary_image * 255

# แสดงภาพโดยใช้ matplotlib
plt.imshow(reconstructed_image, cmap='gray')
plt.title('Reconstructed Image')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# สร้างภาพพื้นหลังสีดำ
image = np.zeros((500, 500, 3), dtype=np.uint8)

# กำหนดจุดของมุมสี่เหลี่ยม
points = [(100, 100), (400, 100), (400, 400), (100, 400)]

# วาดเส้นขอบสี่เหลี่ยม
for i in range(len(points)):
    start_point = points[i]
    end_point = points[(i + 1) % len(points)]  # ใช้ % เพื่อวนลูปกลับไปที่จุดเริ่มต้น
    cv2.line(image, start_point, end_point, (0, 0, 255), thickness=2)

# แสดงภาพใน Colab
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # ไม่ให้แสดงแกน
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# สร้างภาพพื้นหลังสีดำ
image = np.zeros((500, 500, 3), dtype=np.uint8)

# วาดกรอบสี่เหลี่ยมสีแดง
for i in range(100, 400):
    # วาดเส้นแนวตั้ง (เส้นขอบซ้ายและขวา)
    image[i, 100] = [0, 0, 255]  # ขอบซ้าย (สีแดงใน BGR)
    image[i, 400] = [0, 0, 255]  # ขอบขวา (สีแดงใน BGR)

for j in range(100, 401):
    # วาดเส้นแนวนอน (เส้นขอบบนและล่าง)
    image[100, j] = [0, 0, 255]  # ขอบบน (สีแดงใน BGR)
    image[400, j] = [0, 0, 255]  # ขอบล่าง (สีแดงใน BGR)

# แสดงภาพใน Colab
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # ไม่ให้แสดงแกน
plt.show()

```
